6. A new cloumn "COL1_mean_split"

```python
a_list = list(df['Ratio.Per GDP'])
a_mean = mean(a_list)

COL1_mean_split = []
for i in a_list:
    if i < 0.5*a_mean:
        COL1_mean_split.append(0)
    elif i <a_mean:
        COL1_mean_split.append(1)
    else:
        COL1_mean_split.append(2)
#print(COL1_mean_split)

df['COL1_mean_split'] = COL1_mean_split
COL1_mean_split
df.to_csv('emissions_col1_mean_split.csv',index = False)
```

```python
data= pd.read_csv('emissions_col1_mean_split.csv')
sns.countplot(x='COL1_mean_split',data=data)
```
